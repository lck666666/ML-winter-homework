```python
import xlrd
import matplotlib.pyplot as plt
#one scale 2-200-1
#
workbook = xlrd.open_workbook('hdim200.xls') #2-200-200-200-1 Sig 1000sample
sheet1 = workbook.sheet_by_name('page_1')
x=sheet1.col_values(0)
x = x[2:]
y1 = sheet1.col_values(1)
y1 = y1[2:]
sheet2 = workbook.sheet_by_name('page_2')
y2 = sheet2.col_values(1)
y2 = y2[2:]
ax = plt.gca()

#--------------------------------------------------------
workbook = xlrd.open_workbook('hdim5000.xls') #2-200-200-200-200-1 5000sample
sheet3 = workbook.sheet_by_name('page_1')
y3 = sheet3.col_values(1)
y3 = y3[2:]
sheet4 = workbook.sheet_by_name('page_2')
y4 = sheet4.col_values(1)
y4 = y4[2:]
plt.plot(x,y1,'r--',label='Test_1000')
plt.plot(x,y2,'r-',label='Train_1000')
plt.plot(x,y3,'b--',label='Test_5000')
plt.plot(x,y4,'b-',label='Train_5000') 
# ax.set_xscale('log')
ax.set_yscale('log')                
plt.legend(fontsize=18)
plt.title('loss',fontsize=15)
plt.xlabel('epoch',fontsize=15)
plt.savefig('hdimsample.png')
plt.show()
```
